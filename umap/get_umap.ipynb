```python
import h5py

file_path = "/data1011/yuzimu/M2ASDA/data/corrected_tgt.h5ad"

# 以 r+ (读写模式) 打开 h5ad 文件
with h5py.File(file_path, "r+") as f:
    # 顺着报错路径找过去，删掉导致报错的那个坏节点
    if "uns/altExperiments" in f:
        del f["uns/altExperiments"]
        print("清理完成！坏掉的 'altExperiments' 节点已被删除。")
    else:
        print("奇怪，没找到这个节点，我们可能得检查一下别的路径。")
```

```python
import scanpy as sc
corrected_tgt = sc.read_h5ad("/data1011/yuzimu/M2ASDA/data/corrected_tgt.h5ad")
raw_tgt = sc.read_h5ad("/data1011/yuzimu/M2ASDA/data/tgt1_clean.h5ad")


sc.pp.pca(corrected_tgt, random_state=42)
sc.pp.neighbors(corrected_tgt, random_state=42)
sc.tl.umap(corrected_tgt, random_state=42)
sc.pl.umap(corrected_tgt, color=["assay","cell_type"])

sc.pp.pca(raw_tgt, random_state=42)
sc.pp.neighbors(raw_tgt, random_state=42)
sc.tl.umap(raw_tgt, random_state=42)
sc.pl.umap(raw_tgt, color=["assay","cell_type"])
```

```python
import scanpy as sc

# 1. 读取数据
corrected_tgt = sc.read_h5ad("/data1011/yuzimu/M2ASDA/data/corrected_tgt1.h5ad")
raw_tgt = sc.read_h5ad("/data1011/yuzimu/M2ASDA/data/tgt1_clean.h5ad")

# 2. 处理并可视化 Corrected 数据
sc.pp.pca(corrected_tgt, random_state=90)
sc.pp.neighbors(corrected_tgt, random_state=90)
sc.tl.umap(corrected_tgt, random_state=90)
print("Plotting Corrected Data...")
sc.pl.umap(corrected_tgt, color=["assay", "compartment"])

# 3. 处理并可视化 Raw 数据
sc.pp.pca(raw_tgt, random_state=90)
sc.pp.neighbors(raw_tgt, random_state=90)
sc.tl.umap(raw_tgt, random_state=90)
print("Plotting Raw Data...")
sc.pl.umap(raw_tgt, color=["assay", "compartment"])
```
